# Looking at all years (2015 -2022) Indicators Data

```python
import pandas as pd
pd.set_option('display.max_columns', None)
```

```python
df2015 = pd.read_csv('../indicators_clean/clean_indicators_2015.csv')
df2015['Year'] = 2015
df2015.head(3)
```

```python
df2016 = pd.read_csv('../indicators_clean/clean_indicators_2016.csv')
df2016['Year'] = 2016
df2016.head(3)
```

```python
df2017 = pd.read_csv('../indicators_clean/clean_indicators_2017.csv')
df2017['Year'] = 2017
df2017.head(3)
```

```python
df2018 = pd.read_csv('../indicators_clean/clean_indicators_2018.csv')
df2018['Year'] = 2018
df2018.head(3)
```

```python
df2019 = pd.read_csv('../indicators_clean/clean_indicators_2019.csv')
df2019['Year'] = 2019
df2019.head(3)
```

```python
df2020 = pd.read_csv('../indicators_clean/clean_indicators_2020.csv')
df2020['Year'] = 2020
df2020.head(3)
```

```python
df2021 = pd.read_csv('../indicators_clean/clean_indicators_2021.csv')
df2021['Year'] = 2021
df2021.head(3)
```

```python
df2022 = pd.read_csv('../indicators_clean/clean_indicators_2022.csv')
df2022['Year'] = 2022
df2022.head(3)
```

```python
df_all = pd.concat([df2015, df2016, df2017, df2018, df2019, df2020, df2021, df2022], sort = False)
df_all.drop('Unnamed: 0', axis=1, inplace=True)
df_all
```

```python
df_all.Year.value_counts().sort_index()
```

```python
df_all.to_csv('../indicators_clean/clean_indicators_ALL_2015_2022.csv', index=False)
```
