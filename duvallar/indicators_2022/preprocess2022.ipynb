```python
df_categorical = df_categorical.astype('category')
df_categorical.dtypes
```

```python
numerical_columns = list(set(df.columns) - set(df_categorical.columns))
df_numerical = df[numerical_columns]
df_numerical = df_numerical.drop(['IDAY', 'IDATE', 'SEQNO'], axis=1)
df_numerical
```

```python
# Get the total number of columns
total_columns = df.shape[1]

# Calculate the indices to split the columns into eight equal parts
eighth1 = total_columns // 8
eighth2 = 2 * total_columns // 8
eighth3 = 3 * total_columns // 8
eighth4 = 4 * total_columns // 8
eighth5 = 5 * total_columns // 8
eighth6 = 6 * total_columns // 8
eighth7 = 7 * total_columns // 8

# Slice the DataFrame to get the eight equal parts
part1 = df_categorical.iloc[:, :eighth1]
part2 = df_categorical.iloc[:, eighth1:eighth2]
part3 = df_categorical.iloc[:, eighth2:eighth3]
part4 = df_categorical.iloc[:, eighth3:eighth4]
part5 = df_categorical.iloc[:, eighth4:eighth5]
part6 = df_categorical.iloc[:, eighth5:eighth6]
part7 = df_categorical.iloc[:, eighth6:eighth7]
parts = [part1, part2, part3, part4, part5, part6, part7]

print(part1.shape, part2.shape, part3.shape, part4.shape, part5.shape, part6.shape, part7.shape)
total_col = 0
for part in parts:
    total_col += part.shape[1]
print(total_col, df_categorical.shape)
```

```python
index = 0
for part in parts:
    part = pd.get_dummies(part)
    print(part.shape)
    parts[index] = part
    index+=1

```

```python
df_categorical = pd.concat(parts, axis=1)
df_categorical.shape
```

```python
df_categorical.to_parquet("categoricalWithDummies.parquet")
```
