```python
import numpy as np
import pandas as pd 
```

```python
# pd.set_option('display.max_rows', None)
pd.set_option('display.max_columns', None)
```

```python
df = pd.read_parquet("../../../Data/LLCP2022.parquet")
```

```python
df
```

```python
df.replace(to_replace='None', value=np.nan, inplace=True)
df = df.dropna(axis=1, how='all')
```

```python
missing_percentage = (df.isnull().sum() / len(df)) * 100
columns_with_missing_values = missing_percentage[missing_percentage > 50]

result_df = pd.DataFrame({
    'Column': columns_with_missing_values.index,
    'Percentage_of_Missing_Values': columns_with_missing_values.values.round(2)
})
result_df.sort_values(by= 'Percentage_of_Missing_Values', ascending=False)
```

```python
df = df.drop(columns = columns_with_missing_values.index, axis=1)
df
```

```python
missing_percentage = (df.isnull().sum() / len(df)) * 100
columns_with_missing_values = missing_percentage[missing_percentage >= 0]

result_df = pd.DataFrame({
    'Column': columns_with_missing_values.index,
    'Percentage_of_Missing_Values': columns_with_missing_values.values.round(2)
})
# result_df.sort_values(by= 'Percentage_of_Missing_Values', ascending=False)
```

```python
df = df[result_df[result_df['Percentage_of_Missing_Values'] < 10.0].Column.value_counts().index].dropna()
df
```

```python
df._MICHD.value_counts()
```

```python
value_counts_per_column = df.apply(lambda x: x.value_counts())

# Step 2: Check for columns where the unique values are 1.0 and 2.0 only
columns_with_1_and_2 = value_counts_per_column.columns[(value_counts_per_column.loc[1.0] + value_counts_per_column.loc[2.0]).eq(value_counts_per_column.sum(axis=0))]

# Step 3: Display the columns
print("Columns with value counts of 1.0 and 2.0 only:")
print(columns_with_1_and_2)
```

```python
value_counts_per_column = df.apply(lambda x: x.value_counts())
columns_with_4_or_less_unique = value_counts_per_column.columns[value_counts_per_column.nunique() <= 4]

print(columns_with_4_or_less_unique)
```

```python
for col in columns_with_4_or_less_unique:
    print(df[col].value_counts())
```

```python
unique_counts_df = pd.DataFrame({
    'Column Name': df.columns,
    'Number of Unique Values': df.nunique()
})
unique_counts_df.sort_values(by='Number of Unique Values', ascending=False)
```

```python
colsW10OrLessGroups = unique_counts_df[unique_counts_df['Number of Unique Values'] <10]['Column Name']
for col in colsW10OrLessGroups:
    print(df[col].value_counts())

print("Number of Columns with Less than 10 unique values:  ", len(colsW10OrLessGroups))
```

```python
colsW10To20Groups = unique_counts_df[(unique_counts_df['Number of Unique Values'] >= 10) & (unique_counts_df['Number of Unique Values'] < 20)]['Column Name']

for col in colsW10To20Groups:
    print(df[col].value_counts())

print("Number of columns with 10 - 20 unique values:   ", len(colsW10To20Groups))
```
